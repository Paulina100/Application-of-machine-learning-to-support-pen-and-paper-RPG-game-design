```python
import pandas as pd
import os
import pathlib


current_path = os.getcwd()
DATASETS_DIR = pathlib.Path(current_path)
DATASET_FILES = [
    "random_forest_basic_results.csv",
    "random_forest_expanded_results.csv",
    "random_forest_full_results.csv",
]
DATASET_PATHS = [f"{DATASETS_DIR}/{file}" for file in DATASET_FILES]
results = [pd.read_csv(result, index_col=0) for result in DATASET_PATHS]
results = pd.concat(results, join="outer", axis=0).reset_index(drop=True)
```
