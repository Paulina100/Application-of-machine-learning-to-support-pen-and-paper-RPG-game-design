```python
import pandas as pd
import os
import pathlib


current_path = os.getcwd()
DATASETS_DIR = pathlib.Path(current_path)
DATASET_FILES = [
    "lightgbm_basic_results.json",
    "lightgbm_second_results.json",
    "lightgbm_final_results.json",
]
DATASET_PATHS = [f"{DATASETS_DIR}/{file}" for file in DATASET_FILES]
results = [pd.read_json(result) for result in DATASET_PATHS]
results = pd.concat(results, join="outer", axis=0).reset_index(drop=True)
```
