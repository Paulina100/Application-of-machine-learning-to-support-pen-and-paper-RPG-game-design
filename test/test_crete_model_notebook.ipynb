```python
import pandas as pd


data = pd.read_json("../output/bestiary_system_basic.json")
```

```python
data.info()
```

```python
data.head()
```

```python
X = data.drop("level", axis=1)
```

```python
y = data["level"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0, shuffle=True
)
```

```python
from training.create_model import create_model


clf_linear = create_model("train_linear_regression", X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error


y_pred_test = clf_linear.predict(X_test)
y_pred_train = clf_linear.predict(X_train)
rmse_train = mean_squared_error(y_train, y_pred_train, squared=False)
rmse_test = mean_squared_error(y_test, y_pred_test, squared=False)
print(f"RMSE train: {rmse_train:.2f}")
print(f"RMSE test: {rmse_test:.2f}")
```

```python
from sklearn.metrics import accuracy_score
import numpy as np


y_linear_pred = clf_linear.predict(X_test)

# y_linear_pred
result_rf_linear = accuracy_score(y_test, np.around(y_linear_pred, 0), normalize=True)

print(f"Test: {result_rf_linear:.2f}")
```

```python
clf_rf = create_model("train_random_forest", X_train, y_train)
```

```python
print(f"train: {clf_rf.score(X_train, y_train):.2f}")
y_pred_rf = clf_rf.predict(X_test)

result_rf = accuracy_score(y_test, y_pred_rf, normalize=True)

print(f"Test: {result_rf:.2f}")
```
